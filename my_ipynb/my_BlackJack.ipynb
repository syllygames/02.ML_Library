```python
class Card:
    def __init__(self, rank, suit):
        self.suit = suit
        self.rank = rank

    def __str__(self):
        return self.rank + ' of ' + self.suit
```

```python
class Deck:
    """
    Deck is collection of 52 cards.
    """
    suit = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
    rank = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')

    def __init__(self):
        
        d = [Card(r,s) for s in Deck.suit for r in Deck.rank]
        self.deck = d
        
    def __len__(self):
        return len(self.deck)
        
    def deal(self):
        return self.deck.pop()

    def showDeck(self):
        for card in self.deck:
            print(card)
                
    def shuffle(self):
        """
        Next, we want to create a random number we call “r = random.randit(0, i)” we’re passing our range from 0 to i. 
        We only want to pick a number that is to the left our our current position. 
        Because we are starting to the right of our list, 
        iterating backwards and we want to pick a random index that is to the left of that. 
        Then we want to swap two cards, the card at “i” and the card at “r”.
        """
        x = range(len(self.deck) -1, 0, -1)
        for i in x:
            r = random.randint(0,i)
            self.deck[i], self.deck[r] = self.deck[r], self.deck[i] 
            
#         random.shuffle(self.deck)
```

```python
class Player:
    def __init__(self, name, amount = 100, dealer = False):
        self.name = name
        self.bank_roll = amount
        self.isDealer = dealer
        self.hand = []
        self.value = 0
        self.aces = 0       # to track aces.
    
            
    def addCard(self, card):
        self.hand.append(card)
#         value is assigned from the dictionary
        self.value += values[card.rank]
        
        if card.rank == "Ace":
            self.aces += 1
        
        self.adjustForAces()
        
    def adjustForAces(self):
#         print("adjustforAces called !!")
        while self.value >21 and self.aces:
            self.value -= 10
            self.aces -= 1
    
    def showHand(self):
        for card in self.hand:
            print(card)

    def bet(self, val):
        self.bet = val
#         self.bank_roll -= self.bet
        
    def win_bet(self):
        self.bank_roll += self.bet
    
    def lose_bet(self):
        self.bank_roll -= self.bet
        
```

```python
def take_bet(player):
    
    while True:
        try:
            x =int(input('How much would you like to bet? '))
        except ValueError:
            print("sorry, bet must be an integer! ")
        else:
            if player.bank_roll < x:
                print("sorry cant exceed: ", player.bank_roll )
            else:
                player.bet(x)
                break
            
```

```python
def hit(deck, player):
    player.addCard(deck.deal())

    
def hit_or_stand(deck, player):
    global playing 
    
    while True:
        x = input("Hit or Stand? Enter 'h' or 's' ")
        if x[0].lower()== "h":
            hit(deck, player)
        elif x[0].lower() =="s":
            print("player passes. Dealer is playing")
            playing = False
        else:
            print("Try again")
            continue
        break
```

```python
def show_some(player,dealer):
    print("\nDealer's Hand:")
    print(" <card hidden>")
    print('',dealer.hand[1])  
    print("\nPlayer's Hand:", *player.hand, sep='\n ')
    print("Player's Hand =",player.value)
    
def show_all(player,dealer):
    print("\nDealer's Hand:", *dealer.hand, sep='\n ')
    print("Dealer's Hand =",dealer.value)
    print("\nPlayer's Hand:", *player.hand, sep='\n ')
    print("Player's Hand =",player.value)
```

```python
def player_busts(player):
    print("Player busts!")
    player.lose_bet()

def player_wins(player):
    print("Player wins!")
    player.win_bet()

def dealer_busts(player):
    print("Dealer busts!")
    player.win_bet()
    
def dealer_wins(player):
    print("Dealer wins!")
    player.lose_bet()
    
def push(player):
    print("Dealer and Player tie! It's a push.")
```

```python
while True:
    # Print an opening statement
    print('Welcome to BlackJack! Get as close to 21 as you can without going over!\n\
    Dealer hits until she reaches 17. Aces count as 1 or 11.')
    
    # Create & shuffle the deck, deal two cards to each player
    mydeck = Deck()
    mydeck.shuffle()
    
    player1 = Player("P1")
    player1.addCard(mydeck.deal())
    player1.addCard(mydeck.deal())
    
    dealer = Player("D",dealer = True)
    dealer.addCard(mydeck.deal())
    dealer.addCard(mydeck.deal())
            
    # Set up the Player's chips
#     player_chips = Chips()  # remember the default value is 100    
    
    # Prompt the Player for their bet
    take_bet(player1)
    
    # Show cards (but keep one dealer card hidden)
    show_some(player1,dealer)
    
    while playing:  # recall this variable from our hit_or_stand function
        
        # Prompt for Player to Hit or Stand
        hit_or_stand(mydeck,player1) 
        
        # Show cards (but keep one dealer card hidden)
        show_some(player1,dealer) 
        
        # If player's hand exceeds 21, run player_busts() and break out of loop
        if player1.value > 21:
            player_busts(player1)
            break        


    # If Player hasn't busted, play Dealer's hand until Dealer reaches 17 
    if player1.value <= 21:
        
        while dealer.value < 17:
            hit(mydeck,dealer)    
    
        # Show all cards
        show_all(player1, dealer)
        
        # Run different winning scenarios
        if dealer.value > 21:
            dealer_busts(player1)

        elif dealer.value > player1.value:
            dealer_wins(player1)

        elif dealer.value < player1.value:
            player_wins(player1)

        else:
            push(player1)        
    
    # Inform Player of their chips total 
    print("\nPlayer's winnings stand at",player1.bank_roll)
    
    # Ask to play again
    new_game = input("Would you like to play another hand? Enter 'y' or 'n' ")
    
    if new_game[0].lower()=='y':
        playing=True
        continue
    else:
        print("Thanks for playing!")
        break
```
