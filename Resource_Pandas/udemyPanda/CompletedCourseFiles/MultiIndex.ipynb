## The `pivot` Method

```python
sales = pd.read_csv("salesmen.csv", parse_dates = ["Date"])
sales["Salesman"] = sales["Salesman"].astype("category")
# sales.nunique()
sales
```

```python
sales.pivot(index = "Date", columns = "Salesman")
```

```python
sales.pivot(index = "Date", columns = "Salesman", values= "Revenue")
```

## The `pivot_table()` Method

```python
foods =pd.read_csv("foods.csv")
foods.head(3)
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], aggfunc = "min")
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], aggfunc = "mean")
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], aggfunc = "sum")
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], columns = "City", aggfunc = "min").head(3)
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], columns = [ "Frequency", "City"], aggfunc = "sum")
```

```python
# Pivot_table method is also directly available on pandas.
pd.pivot_table(data = foods, values = "Spend", index = ["Gender", "Item"], columns = "City", aggfunc = "min").head(3)
```

## The `pd.melt()` Method

```python
#  Essentially reverse method of pivot table.

sales = pd.read_csv("quarters.csv")
sales
```

```python
df2 =pd.melt(sales, id_vars = "Salesman", var_name = "Quarter", value_name = "Revenue")
df2
```

```python
df2.pivot(index = "Salesman", columns="Quarter", values = "Revenue")
df2
```

```python
df2.pivot_table(index = "Salesman", columns="Quarter", values = "Revenue")
```
